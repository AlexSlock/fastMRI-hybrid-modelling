```python
import glob
import h5py

path_to_test_mri_data = '/usr/local/micapollo01/MIC/DATA/SHARED/NYU_FastMRI/multicoil_test_full/'
file_paths_test_GT = sorted(glob.glob(path_to_test_mri_data+"file_brain_*"))

enough_coils = []

for file in file_paths_test_GT:
    with h5py.File(file,'r') as f:
        k = f['kspace'][()]
        dims = k.shape
        if dims[1] > 12:
            if dims[2] == 640 and dims[3] == 320:
                enough_coils.append(file)
            else:
                print(str(file)+' has enough coils, but not the right slice dimensions.')
        else:
            print(str(file)+' has not enough coils.')

```
