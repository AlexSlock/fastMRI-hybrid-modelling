```python
import os
import h5py
import numpy as np
import matplotlib.pyplot as plt

download_path = '/usr/local/micapollo01/MIC/DATA/SHARED/NYU_FastMRI/Preprocessed'
test_path = os.path.join(download_path,'multicoil_test')

# R = 4
file_name = 'file_brain_AXFLAIR_209_6001340.h5'
hf = h5py.File(os.path.join(test_path, file_name),'r')

volume = hf['sense_data'][()]
slice = volume[3] # Choosing the third slice of this volume
print(slice.dtype)
print(slice.shape)
```

```python
plt.imshow(np.abs(slice), cmap='gray')
```

```python
volume = hf['cs_data'][()]
slice = volume[3] # Choosing the third slice of this volume
print(slice.dtype)
print(slice.shape)
```

```python
plt.imshow(np.abs(slice), cmap='gray')
```

```python
input_kspace = hf['grappa_data'][()]

import fastmri
from fastmri.data import transforms as T
input_kspace_T = T.to_tensor(input_kspace)    
sampled_image_T = fastmri.ifft2c(input_kspace_T)
sampled_image_abs = fastmri.complex_abs(sampled_image_T)
sampled_image_rss = fastmri.rss(sampled_image_abs, dim=1)

slice = sampled_image_rss[3,:,:].numpy()
```

```python
plt.imshow(np.abs(slice), cmap='gray')
```
