```python
import os
import pickle

def compare_lists(list1, list2):
    # Check if the lists have different lengths
    if len(list1) != len(list2):
        return "Different lengths"

    # Compare each pair of dictionaries in the lists
    for d1, d2 in zip(list1, list2):
        # Check if the dictionaries have different keys
        keys1 = list(d1.keys())
        keys2 = list(d2.keys())
        if keys1 != keys2:
            return "Different keys"

    return "Same"

def compare_batches(prefix1, prefix2, directory):
    # Get the list of batch files
    batches1 = [f for f in os.listdir(directory) if f.startswith(prefix1)]
    batches2 = [f for f in os.listdir(directory) if f.startswith(prefix2)]

    # Sort the batch files by batch number
    batches1.sort(key=lambda f: int(os.path.splitext(f)[0].split('_')[-1]))
    batches2.sort(key=lambda f: int(os.path.splitext(f)[0].split('_')[-1]))

    print(batches1)
    print(batches2)

    # Compare each pair of batches
    for batch1, batch2 in zip(batches1, batches2):
        # Load the lists from the files
        with open(os.path.join(directory, batch1), 'rb') as f:
            list1 = pickle.load(f)
        with open(os.path.join(directory, batch2), 'rb') as f:
            list2 = pickle.load(f)

        # Compare the lists
        result = compare_lists(list1, list2)
        if result != "Same":
            # Get the batch numbers from the file names
            batch_num1 = int(os.path.splitext(batch1)[0].split('_')[-1])
            batch_num2 = int(os.path.splitext(batch2)[0].split('_')[-1])

            # Print the batch numbers and the result
            print(f"Batches {batch_num1} and {batch_num2} have {result}")


```

```python
grappa_path = '/usr/local/micapollo01/MIC/DATA/STUDENTS/mvhave7/Results/Preprocessing/Backlog/grappa/'

compare_batches('grappa_p_batch_', 'grappa_wt_batch_', grappa_path)
```
